```python
import pandas as pd
```

```python
df_energy = pd.read_csv("/Users/luizbeethoven/Documents/CSV's/municipios_output.csv")
df_energy.head()
```

```python
df_energy = df_energy.rename(columns={
    "Categoria": "Categoria de uso",
    "Alta": "Alta tensão (>1000V)",
    "Baixa": "Baixa tensão (<1000V)"
})
```

```python
df_energy.head()
```

```python
df_energy["Município"] = df_energy["Município"].str.strip()
df_energy["Categoria de uso"] = df_energy["Categoria de uso"].str.strip()

# build a formatted text block per municipality
def build_block(group):
    lines = []
    for _, row in group.iterrows():
        lines.append(
            f"{row['Categoria de uso']}: "
            f"Alta={row['Alta tensão (>1000V)']:,} | "
            f"Baixa={row['Baixa tensão (<1000V)']:,} | "
            f"Auto={row['Autoconsumo']:,} | "
            f"Total={row['Total']:,}"
        )
    return "\n".join(lines)

energy_block = df_energy.groupby("Município").apply(build_block).reset_index()
energy_block.columns = ["Município", "Energia (kWh) por categoria"]
```
